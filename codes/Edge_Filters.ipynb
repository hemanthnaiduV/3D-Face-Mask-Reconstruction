```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image(img):
    # Resize and blur for consistency
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5, 5), 1.5)
    return blur

def fuse_edges(image):
    # Canny
    edges_canny = cv2.Canny(image, 50, 150)

    # Sobel (both directions)
    sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    edges_sobel = cv2.convertScaleAbs(np.hypot(sobelx, sobely))

    # Laplacian
    laplacian = cv2.Laplacian(image, cv2.CV_64F)
    edges_lap = cv2.convertScaleAbs(laplacian)

    # Scharr
    scharrx = cv2.Scharr(image, cv2.CV_64F, 1, 0)
    scharry = cv2.Scharr(image, cv2.CV_64F, 0, 1)
    edges_scharr = cv2.convertScaleAbs(np.hypot(scharrx, scharry))

    # Weighted fusion (you can tune these weights)
    fused = 0.4 * edges_canny + 1 * edges_sobel + 0.2 * edges_lap + 0.2 * edges_scharr
    fused = np.clip(fused, 0, 255).astype(np.uint8)

    # Optional: binarize
    _, final = cv2.threshold(fused, 30, 255, cv2.THRESH_BINARY)

    return {
        'Canny': edges_canny,
        'Sobel': edges_sobel,
        'Laplacian': edges_lap,
        'Scharr': edges_scharr,
        'Fused': final
    }

def visualize_all(edges_dict):
    plt.figure(figsize=(15, 8))
    for i, (title, edge_map) in enumerate(edges_dict.items()):
        plt.subplot(2, 3, i+1)
        plt.imshow(edge_map, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# === Main Code (Colab-friendly) ===
from google.colab import files
uploaded = files.upload()  # Upload your image here

for fn in uploaded.keys():
    image = cv2.imread(fn)
    processed = preprocess_image(image)
    edges = fuse_edges(processed)
    visualize_all(edges)
```

```python
!pip install mediapipe --quiet

import cv2
import numpy as np
import mediapipe as mp
import matplotlib.pyplot as plt

# Initialize MediaPipe Face Mesh
mp_face_mesh = mp.solutions.face_mesh
face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True)

# Utility function to extract unique landmark indices from frozenset
def extract_indices(frozenset_pairs):
    return sorted(set(i for pair in frozenset_pairs for i in pair))

# Facial feature landmark indices
FACIAL_FEATURES = {
    'lips': extract_indices(mp_face_mesh.FACEMESH_LIPS),
    'left_eye': extract_indices(mp_face_mesh.FACEMESH_LEFT_EYE),
    'right_eye': extract_indices(mp_face_mesh.FACEMESH_RIGHT_EYE),
    'face_outline': extract_indices(mp_face_mesh.FACEMESH_FACE_OVAL),
    'nose': [1, 2, 98, 327, 168, 94, 4, 5, 6, 195, 197, 3]  # includes bridge and nostrils
}

def get_feature_mask(image, landmarks, feature_ids):
    mask = np.zeros(image.shape[:2], dtype=np.uint8)
    h, w = image.shape[:2]
    points = [(int(landmarks[i].x * w), int(landmarks[i].y * h)) for i in feature_ids]
    if points:
        cv2.fillPoly(mask, [np.array(points, dtype=np.int32)], 255)
    return mask

def apply_feature_edges(image, landmarks):
    masks = {}
    for name, ids in FACIAL_FEATURES.items():
        masks[name] = get_feature_mask(image, landmarks, ids)

    blur_gray = cv2.GaussianBlur(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), (5, 5), 1.5)
    edges = {}

    for name, mask in masks.items():
        masked_img = cv2.bitwise_and(blur_gray, blur_gray, mask=mask)
        edge = cv2.Canny(masked_img, 50, 150)
        edges[name] = edge

    return edges

def visualize_feature_edges(image, edge_maps):
    overlay = np.zeros_like(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))
    for edge in edge_maps.values():
        overlay = cv2.bitwise_or(overlay, edge)

    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(overlay, cmap='gray')
    plt.title("Facial Feature Edges")
    plt.axis("off")

    plt.show()

# === Upload and Process Image ===
from google.colab import files
uploaded = files.upload()

for fn in uploaded:
    image = cv2.imread(fn)
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    result = face_mesh.process(rgb)

    if result.multi_face_landmarks:
        for face_landmarks in result.multi_face_landmarks:
            edge_maps = apply_feature_edges(image, face_landmarks.landmark)
            visualize_feature_edges(image, edge_maps)
    else:
        print("No face detected.")
```
